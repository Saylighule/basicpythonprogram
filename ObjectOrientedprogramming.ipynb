```python
# static method
```

```python
class calculator:
    @staticmethod
    def add(a,b):
        return a+b
```

```python
calculator.add(10,20)
```

```python
class calculator:
    @staticmethod
    def mul(a,b):
        return a*b
```

```python
calculator.mul(10,4)
```

```python
# Encapsulation
```

```python
class BankAccount:

    def __init__(self,balance):
        self.__balance=balance

    def get_balance(self):
        return self.__balance
```

```python
BankAccount(1000).get_balance()
```

```python
#method overriding 
```

```python
class vehicle:
    def start(self):
        print("vehicle is starting")

class car(vehicle):
    def start(self):
        print("car is starting")
```

```python
c=car()
```

```python
c.start()
```
