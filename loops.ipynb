```python
for i in range(1,11):
    
    if i == 5:
        break

    print(i)
```

```python
for i in range(1,11):
    
    if i == 5:
        continue

    print(i)
```

```python
n = int(input("Enter a positive integer:"))
factorial = 1
for i in range(1,n + 1):
    factorial*= i
    print(f"The factorial of {n} is {factorial}").
```
