```python
#6.	Write a Python program to calculate the factorial of a number, count its digits, and display the reverse of the number.
```

```python
a = int(input("Enter a number: "))

# Factorial
fact = 1
for i in range(1, a+ 1):
    fact *= i

# Count digits
digits = len(str(a))

# Reverse number
reverse = str(a)[::-1]

print("Factorial =", fact)
print("Number of digits =", digits)
print("Reverse =", reverse)
```

```python
#11.Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
word = "Python"
```

```python
# Reverse a string
```

```python
print(word[::-1])
```

```python
# Count the vowels
```

```python
count = 0
for ch in word.lower():
    if ch in "aeiou":
        count += 1
        print("Vowels =",count)
```

```python
# Count the consonents
```

```python
count = 0
for ch in word.lower():
    if ch.isalpha() and ch not in "aeiou":
        count += 1
        print("Consonants =", count)
```
